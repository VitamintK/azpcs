```python
def get_possibilities(EXP, TERMS, IGNORE_SMALLEST=0):
    possibilities = []
    pows = []

#     TERMS = 25
#     EXP = 16

    for i in range(TERMS+1):
        pows.append(pow(i,EXP))
    for i in range(pow(2,TERMS-IGNORE_SMALLEST)):
        x = i
        b = 0
        ans = 0
        j = IGNORE_SMALLEST
        while x != 0:
            if x&1:
                ans += pows[j]
            x >>= 1
            j += 1
        possibilities.append((bin(i), ans))
    return possibilities
```
